```python
import math
import matplotlib.pyplot as plt
import networkx as nx
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Subplot 1
G = nx.binomial_graph(30, 0.07)
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, node_size=50, with_labels=False, ax=axes[0])
axes[0].set_title("Network containing 'giant component'")

# Subplot 2
G = nx.complete_graph(3, create_using=nx.Graph)
G.add_edge(1, 1)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos, node_size=50, with_labels=False, ax=axes[1])
axes[1].set_title('Small network with self-loop')
```
