# Setup

```python
import matplotlib.pyplot as plt


plt.style.use('seaborn')
```

# Results

```python
age_distribution = {
    '0-9': 11.24,
    '10-19': 11.53,
    '20-34': 20.43,
    '35-49': 19.30,
    '50-64': 19.33,
    '65-79': 13.43,
    '80+': 4.74
}
```

```python
plt.figure(figsize=(7, 4))
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age bracket')
plt.ylabel('Percentage of Population')

bars = plt.bar(age_distribution.keys(), age_distribution.values())
plt.bar_label(bars)

plt.ylim(0, 25)

plt.show()
```
