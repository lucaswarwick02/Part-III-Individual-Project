# Setup

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as stats


plt.style.use('seaborn')
```

```python
def generate_distribution (mu: float, variance: float, N=100):
    sigma = math.sqrt(variance)
    
    X = np.linspace(0, 10, N)
    
    return X, stats.norm.pdf(X, mu, sigma)
```

# Results

```python
data = {
    'High': {
        'X': [],
        'y': [],
        'mu': 5,
        'variance': 1,
        'color': 'red'
    },
    'Medium': {
        'X': [],
        'y': [],
        'mu': 5,
        'variance': 3,
        'color': 'green'
    },
    'Low': {
        'X': [],
        'y': [],
        'mu': 5,
        'variance': 25,
        'color': 'blue'
    },
}
```

```python
plt.figure(figsize=(7, 4))

for key in data.keys():
    X, y = generate_distribution(mu=data[key]['mu'], variance=data[key]['variance'])
    data[key]['X'] = X
    data[key]['y'] = y
    plt.plot(X, y, color=data[key]['color'], label=key)
    
plt.legend()
plt.show()
```

```python
for key in data.keys():
    print(f'Type: {key}')
    print(f'... Variance: {np.var(data[key]["y"])}')
    print(f'... Standard Deviation: {np.std(data[key]["y"])}')
```

Standard deviation seems to be an appropriate measure for 'flatness' of the curves, therefore after each simulation has finished it's iterations, it would be useful to record the standard deviation for 'infections', 'hospitalisation' and 'deaths'. It can then be averaged alongside the other operations during the aggregation stage to output a final value for that runthrough.
