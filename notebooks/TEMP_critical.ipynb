```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import re


plt.style.use('seaborn')
```

```python
def plot_infections (run_folder: str):
    run_folder_path = os.path.join(os.path.abspath('../'), 'out', run_folder)

    results = []

    for filename in os.listdir(run_folder_path):
        filename_path = os.path.join(run_folder_path, filename)
        match = re.match(r'RandomOneOff\(\d+\,\s(\d+\.\d+),\s(\d+\.\d+)\)', filename)
        
        if match is None:
            continue
        
        beta = float(match.groups()[1])
        
        totals_df = pd.read_csv(os.path.join(filename_path, 'totals.csv'))
        
        results.append((beta, totals_df.iloc[-1]['Infected']))
        
        
    plt.figure(figsize=(7, 4))

    plt.plot([x for x, y in results], [y for x, y in results])

    plt.show()
```

```python
plot_infections('184650_27022023')
```
